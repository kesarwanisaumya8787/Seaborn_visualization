```python
#  Create a scatter plot to visualize the relationship between two variable, by generating using a synthetic dataset.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
x = np.random.randint(1,100,40)
y =np.random.normal(0,10,40) 
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot of X vs Y')
plt.show()
```

```python
#    Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = np.random.normal(50,10,size=1000)  

sns.histplot(data, kde=True)
# Display the plot
plt.title('Distribution of Randomly Generated Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
